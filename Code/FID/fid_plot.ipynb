```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
models = ['Original vs DDPM', 'Original vs PGGANs 150k', 'Original vs PGGANs 160k']
normal_fid = [110.62, 135.71, 107.57]  # FID values for NORMAL class
pneumonia_fid = [113.72, 191.78, 145.30]  # FID values for PNEUMONIA class

#Full data
#models = ['Original vs DDPM', 'Original vs PGGANs 150k', 'Original vs PGGANs 160k']
#normal_fid = [105.00, 125.84, 95.60]  # FID values for NORMAL class
#pneumonia_fid = [106.80, 192.57, 132.39]  # FID values for PNEUMONIA class

# Set up the bar plot
width = 0.35  # Width of each bar
x = range(len(models))

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for NORMAL class
rects1 = ax.bar([i - width/2 for i in x], normal_fid, width, label='NORMAL')

# Plot bars for PNEUMONIA class
rects2 = ax.bar([i + width/2 for i in x], pneumonia_fid, width, label='PNEUMONIA')

# Add labels, title, and legend
ax.set_xlabel('Models')
ax.set_ylabel('FID (Frchet Inception Distance)')
ax.set_title('FID Score for Original (all data) vs Generated Images ')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Add values on top of each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

# Show the plot
plt.tight_layout()
output_file = 'fid_comparison_plot.png'
plt.savefig(output_file)
plt.show()
```
